# insert into hospital

```python
print("Let's input the hospital staff information!")
while True:
    staff_id = input('staff\'s id: ')
    department = input('Department\'s name: ')
    sql = '''insert into hospital
             (staff_id, department)
             values
             (:id, :department)'''
             # These values are "named parameters" (like place holders)
             # Tells the SQLite library that something will be substitued here
    
    # Use the cursor to execute the statement
    # Here, a dictionary has been added of the named parameters and the items
    # to be inserted.
    cursor.execute(sql, {'id':staff_id, 'department':department})
    
    # Commit. Telling SQLite to save the new data. The data would be lost otherwise.
    conn.commit()
    
    # See if the user wants to add another student's information into the database...
    cont = input("Another staff? ")
    # If no, break (break out of the while loop) and close the connection.
    # Otherwise, go to the beginning of the while loop to add another student
    if cont[0].lower() == 'n':
        break
```

# Query 1

Find the doctor who has been working for 10 years

```python
query1 = """
SELECT staff_id, name, working_years
FROM doctors 
WHERE working_years = '10' 
"""
```

```python
pd.read_sql_query(query1, conn)
```

# Query 2

Find the doctor who has been working for longest years in each department and make the descending order by years.

```python
query2 = """
SELECT hospital.staff_id, name, working_years, department
FROM hospital
NATURAL JOIN doctors
GROUP BY department
HAVING max(working_years)
ORDER BY working_years DESC
"""
```

```python
pd.read_sql_query(query2, conn)
```

# Query 3

Find the patients who have the age over 60, descendent ordered by age.

```python
query3 = """
SELECT patient_id, patient_name, age
FROM patients
WHERE age > 60
ORDER BY age DESC
"""
```

```python
pd.read_sql_query(query3, conn)
```

# Query 4

Find the patients who have taken covid test age and the result is positive.

```python
query4 = """
SELECT patient_id, patient_name, test_name, result
FROM patients 
NATURAL JOIN tests
WHERE result = "positive"
"""
```

```python
pd.read_sql_query(query4, conn)
```
