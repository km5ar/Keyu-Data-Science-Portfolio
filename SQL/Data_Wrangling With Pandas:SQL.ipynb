## Problem 16
Write `pandas` code that replicates the output of the following SQL code:
```
SELECT gender_balance, premiums, COUNT(*)
FROM whpps
GROUP BY gender_balance, premiums
HAVING gender_balance is NOT NULL and premiums is NOT NULL;
```
[2 points]

````python
df_16 = df_12.loc[:,['gender_balance', 'Insurance_offer_fulltime_employees']]
df_16.dropna(subset=['gender_balance','Insurance_offer_fulltime_employees'], inplace=True)
df_16.groupby(['gender_balance', 'Insurance_offer_fulltime_employees'])['gender_balance'].count()
````

````python
myquery="""
SELECT gender_balance, Insurance_offer_fulltime_employees, COUNT(*)
FROM whpps
GROUP BY gender_balance, Insurance_offer_fulltime_employees
HAVING gender_balance is NOT NULL and Insurance_offer_fulltime_employees is NOT NULL;
"""
````

````python
whpps_cursor.execute(myquery)
whpps_df = whpps_cursor.fetchall()
pd.DataFrame(whpps_df)
````
